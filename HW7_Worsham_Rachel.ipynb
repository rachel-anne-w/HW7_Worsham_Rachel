```r
#Creates data frame
q1 <- data.frame(ID = c(1,2,3,4,5), Name = c('Peter','Amy', 'Ryan', 'Gary', 'Michelle'), Salary = c(623.3, 515.2, 611, 729, 843.25))
q1
```

```r
#adds department column
Department <-  c('IT','Education', 'Finance', 'Music Producer','Bank Manager' )
q1$Department<-Department
q1
```

```r
qe = q1[c(1,4,5),c('Salary')]
barplot(qe, main='Salary Comparison', names.arg=c('Peter','Ryan','Michelle'))


```

```r
min(q1$Salary)
max(q1$Salary)
median(q1$Salary)
```

```r
Salaries<- c(515.2,843.25,623.3)
names = c('Amy', 'Michelle','Peter')
colors = c('pink','purple','yellow')
pie(Salaries, label=Salaries, col = colors)
legend('bottomright',names, fill = colors )
```
